```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'SampleSuperstore.csv'  # Update the path to your file
data = pd.read_csv(file_path)
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a figure with multiple subplots
fig, axes = plt.subplots(3, 2, figsize=(20, 24))

# Plot 1: Distribution of Sales
sns.histplot(data['Sales'], bins=50, kde=True, color='blue', ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Sales')
axes[0, 0].set_xlabel('Sales')
axes[0, 0].set_ylabel('Frequency')

# Plot 2: Distribution of Profit
sns.histplot(data['Profit'], bins=50, kde=True, color='green', ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Profit')
axes[0, 1].set_xlabel('Profit')
axes[0, 1].set_ylabel('Frequency')

# Group by region and calculate total sales and profit
region_analysis = data.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

# Plot 3: Total Sales by Region
sns.barplot(x='Region', y='Sales', data=region_analysis, palette='viridis', ax=axes[1, 0])
axes[1, 0].set_title('Total Sales by Region')
axes[1, 0].set_xlabel('Region')
axes[1, 0].set_ylabel('Total Sales')

# Plot 4: Total Profit by Region
sns.barplot(x='Region', y='Profit', data=region_analysis, palette='viridis', ax=axes[1, 1])
axes[1, 1].set_title('Total Profit by Region')
axes[1, 1].set_xlabel('Region')
axes[1, 1].set_ylabel('Total Profit')

# Group by category and sub-category and calculate total sales and profit
category_analysis = data.groupby(['Category', 'Sub-Category'])[['Sales', 'Profit']].sum().reset_index()

# Plot 5: Total Sales by Category and Sub-Category
sns.barplot(x='Sales', y='Sub-Category', hue='Category', data=category_analysis, palette='viridis', ax=axes[2, 0])
axes[2, 0].set_title('Total Sales by Category and Sub-Category')
axes[2, 0].set_xlabel('Total Sales')
axes[2, 0].set_ylabel('Sub-Category')
axes[2, 0].legend(title='Category', bbox_to_anchor=(1, 1))

# Plot 6: Impact of Discount on Profit
sns.scatterplot(x='Discount', y='Profit', data=data, hue='Category', palette='viridis', alpha=0.6, ax=axes[2, 1])
axes[2, 1].set_title('Impact of Discount on Profit')
axes[2, 1].set_xlabel('Discount')
axes[2, 1].set_ylabel('Profit')
axes[2, 1].legend(title='Category', bbox_to_anchor=(1, 1))

# Adjust layout for better appearance
plt.tight_layout()

# Show the dashboard
plt.show()
```
